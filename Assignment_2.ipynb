## Question 1: Age Group Classification

Write a Python program that takes the age of a person as input and classifies them into one of the following age groups:
"Child" if the age is between 0 and 12 (inclusive).
"Teenager" if the age is between 13 and 19 (inclusive).
"Adult" if the age is between 20 and 64 (inclusive).
"Senior" if the age is 65 or older.

```python
age = float(input ("Enter your age: "))
print ("Your age is" , age , "years.")

if age >= 0 and age <= 12:
  print ("You are a Child.")
elif age >= 12 and age <= 19:
  print ("You are a Teenager.")
elif age >= 19 and age <= 64:
  print ("You are an Adult.")
else:
  print ("You are a Senior Citizen")
```

## Question 5.
Write a program that replaces spaces with underscores, counts how many underscores are there, and finds the position of the first underscore.
Use text: "I'm learning Python programming"?

## Question 6:

Write a Python program that:
Takes a string as input.
Prints all characters of the string except the last 3 using slicing.
Prints every second character of the string using slicing.
Prints the string in reverse using slicing

```python
text = input ("Write any phrase: ")
print ("Your given phrase is: " , text)

text1 = text[:-3]
print ("Your given phrase after dropping last 3 characters: " , text1)

text2 = text[1::2]
print ("Every 2nd character of given phrase: ", text2)

text3 = text [::-1]
print ("Your given phrase in reverse order: " , text3)
```

## Question 7.
Given a sentence, you need to reverse the order of the words in the sentence while preserving their original capitalization and spaces. Use text : "Hello World, We are learning Python"

```python
text = "Hello World, We are learning Python"
print ("Original phrase is: " , text)

text1 = text.split()[::-1]
t = []
for i in text1:
    t.append(i)
print("Given phrase in reverse order: ", " ".join(t))

text2 = text[::-1]
print ("Given phrase in backword order: " , text2)
```

## Question 10: Find the Longest Word

Write a Python program that finds the longest word in a given string.
Example:
Input: "I love programming in Python" Output: "programming"

```python
sentence = "I love programming in Python."
print ("Original sentence is: ", sentence)

words = sentence.split()
longest_word = ""

for i in words:
    if len(words) > len(longest_word):
         longest_word = i

print ("Longest word in the original sentence is: ", longest_word)
```
